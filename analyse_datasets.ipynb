## Etude de l'échantillon de test (non variable)
### Analyse descriptive des variables

```python
df_test = pd.read_csv('test_set.csv')
df_test.head()
```

```python
cities = np.unique(df_test['city'])
for city in cities:
    print('Number of occurence of ', city,' in test set = ' "{:.2f}".format(len(df_test[df_test['city'] == city].index)/len(df_test.index)*100), "%")

percent_city_df_test = ["{:.2f}".format(len(df_test[df_test['city'] == city].index)/len(df_test.index)*100) for city in cities]

cit = np.array([len(df_test[df_test['city'] == i].index) for i in cities])
plt.pie(cit, labels = cities, startangle = 90)
plt.show()
```

```python
languages = np.unique(df_test['language'])
for language in languages:
    print('Number of occurence of ', language,' in test set = ' "{:.2f}".format(len(df_test[df_test['language'] == language].index)/len(df_test.index)*100), "%")

percent_lan_df_test = ["{:.2f}".format(len(df_test[df_test['language'] == language].index)/len(df_test.index)*100) for language in languages]

lan = np.array([len(df_test[df_test['language'] == i].index) for i in languages])
plt.pie(lan, labels = languages, startangle = 90)
plt.show()
```

```python
mobiles = np.unique(df_test['mobile'])
for mobile in mobiles:
    print('Number of occurence of ', mobile,' in test set = ', len(df_test[df_test['mobile'] == mobile].index))

mob = np.array([len(df_test[df_test['mobile'] == i].index) for i in mobiles])
plt.pie(mob, labels = mobiles, startangle = 90)
plt.show()    
```

```python
sns.distplot(df_test['date'],kde=True).set(title="Date",xlabel=None)
plt.show()
```

```python
for language in languages:
    print("LANGUAGE = ", language, "\n")
    df_lan = df[df['language'] == language]
    df_lan.reset_index()
    df_test_lan = df_test[df_test['language'] == language]
    df_test_lan.reset_index()
    for city in cities:
        print("CITY = ", city)
        print("Nombre pour le dataset de test = ", "{:.2f}".format(len(df_test_lan[df_test_lan['city']== city].index)/len(df_test_lan.index)*100), "%")
        print("Nombre pour le dataset des requêtes = ", "{:.2f}".format(len(df_lan[df_lan['city']== city].index)/len(df_lan.index)*100), "%")

    print('----------------------------------\n')
```

```python
for mobile in mobiles:
    print("MOBILE = ", mobile, "\n")
    df_lan = df[df['mobile'] == mobile]
    df_lan.reset_index()
    df_test_lan = df_test[df_test['mobile'] == mobile]
    df_test_lan.reset_index()
    for city in cities:
        print("CITY = ", city)
        print("Nombre pour le dataset de test = ", "{:.2f}".format(len(df_test_lan[df_test_lan['city']== city].index)/len(df_test_lan.index)*100), "%")
        print("Nombre pour le dataset des requêtes = ", "{:.2f}".format(len(df_lan[df_lan['city']== city].index)/len(df_lan.index)*100), "%")

    print('----------------------------------\n')
```

```python
for language in languages:
    print("LANGUAGE = ", language, "\n")
    df_lan = df[df['language'] == language]
    df_lan.reset_index()
    df_test_lan = df_test[df_test['language'] == language]
    df_test_lan.reset_index()
    for city in cities:
        print("CITY = ", city)
        df_test_city_lan = df_test_lan[df_test_lan['city']== city]
        df_city_lan = df_lan[df_lan['city']== city]
        for mobile in mobiles:
            print("MOBILE = ", mobile)
            print("Nombre pour le dataset de test = ", "{:.2f}".format(len(df_test_city_lan[df_test_city_lan['mobile']== mobile].index)/len(df_test_lan.index)*100), "%")
            print("Nombre pour le dataset des requêtes = ", "{:.2f}".format(len(df_city_lan[df_city_lan['mobile']== mobile].index)/len(df_lan.index)*100), "%")

    print('----------------------------------\n')
```
