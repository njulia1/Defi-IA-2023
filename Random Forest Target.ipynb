# Chargement des donn√©es

```python
data = pd.read_csv('data3.csv')
data = data.drop(columns = 'hotel_id')
data = data.drop(columns = 'avatar_name')
data.head()
```

```python
data
```

## on encode les variables par le prix moyen suivant les crit√®res

```python
data["city_encoded"] = data.groupby("city")["price"].transform("mean")
data["group_encoded"] = data.groupby("group")["price"].transform("mean")
data["brand_encoded"] = data.groupby("brand")["price"].transform("mean")
data["language_encoded"] = data.groupby("language")["price"].transform("mean")
```

```python
data
```

```python
data_city = data.sort_values(by=['city'])
cities = data['city'].to_numpy()
cities_encoded = data['city_encoded'].to_numpy()
dictionnary_city = dict(zip(cities, cities_encoded))

data_language = data.sort_values(by=['language'])
language = data['language'].to_numpy()
language_encoded = data['language_encoded'].to_numpy()
dictionnary_language = dict(zip(language, language_encoded))

data_group = data.sort_values(by=['group'])
group = data['group'].to_numpy()
group_encoded = data['group_encoded'].to_numpy()
dictionnary_group = dict(zip(group, group_encoded))

data_brand = data.sort_values(by=['brand'])
brand = data['brand'].to_numpy()
brand_encoded = data['brand_encoded'].to_numpy()
dictionnary_brand = dict(zip(brand, brand_encoded))
```

```python
data = data.drop(columns = 'group')
data = data.drop(columns = 'city')
data = data.drop(columns = 'language')
data = data.drop(columns = 'brand')
```

```python
data
```

## Random Forest

```python
# On d√©finit la fonction MAPE : Mean Absolute Percentage Error
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

```python
#S√©paration du jeu de donn√©es en jeu d'apprentissage et de test
price = data["price"]
price = np.log(price)
X_train, X_test, price_train, price_test = train_test_split(data,price,test_size=0.25,random_state=11)
X_train = X_train.drop(['price'],axis=1)
X_test = X_test.drop(['price'],axis=1)
X_train.shape, price_train.shape
```

```python
rf= RandomForestRegressor(n_estimators=400,max_features=2)
rfFit=rf.fit(X_train, price_train)
```

```python
predictions = rf.predict(X_test)
```

```python
print("MSE=",mean_squared_error(price_test,predictions))
print("R2=",r2_score(price_test,predictions))
print("MAPE=", MAPE(price_test,predictions))
```
